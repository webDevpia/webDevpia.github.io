```python
import os
from dotenv import load_dotenv
load_dotenv()  # .env 파일의 환경변수 로드

# API 키와 환경명 가져오기
pinecone_api_key = os.getenv("PINECONE_API_KEY")
pinecone_env = os.getenv("PINECONE_ENVIRONMENT")
```

```python
# Pinecone 연결 및 인덱스 초기화
import pinecone
from pinecone import Pinecone as PineconeClient
from pinecone import ServerlessSpec

# Pinecone 클라이언트 초기화
pc = PineconeClient(api_key=pinecone_api_key)
pc.create_index(name="langchain-rag",
                dimension=1536,
                metric="cosine",
                spec=ServerlessSpec(cloud="aws",
                                    region="us-east-1"))
```

```python
# OpenAI 임베딩 모델 설정 (text-embedding-3-small 사용)
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
from langchain_core.documents import Document

# 예시 문서 생성 (내용과 메타데이터)
doc1 = Document(page_content="Building an exciting new project with LangChain - come check it out!",
                metadata={"source": "tweet"})
doc2 = Document(page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
                metadata={"source": "news"})
doc3 = Document(page_content="LangGraph is the best framework for building stateful, agentic applications!",
                metadata={"source": "tweet"})

# 벡터스토어 초기화 (PineconeVectorStore에 Pinecone 인덱스와 임베딩 객체 연결)
from langchain_pinecone import PineconeVectorStore
vector_store = PineconeVectorStore(index_name="langchain-rag", embedding=embeddings)

# 문서들을 벡터 임베딩하여 Pinecone에 저장
vector_store.add_documents([doc1, doc2, doc3])
```

```python
import pandas as pd

# 예시를 위한 간단한 데이터프레임 생성 (실제로는 CSV를 로드)
# df = pd.read_csv("KoreanDramas.csv")  # 실제 데이터셋 사용 시
data = [
    {
        "title": "응답하라 1988",
        "year": 2015,
        "genre": ["드라마", "코미디"],
        "director": "신원호",
        "actors": ["혜리", "박보검", "류준열"],
        "rating": 9.2,
        "synopsis": "1988년 서울, 쌍문동 이웃들 사이의 우정과 가족애를 그린 드라마."
    },
    {
        "title": "기생충",
        "year": 2019,
        "genre": ["드라마", "스릴러"],
        "director": "봉준호",
        "actors": ["송강호", "이선균", "조여정"],
        "rating": 8.6,
        "synopsis": "가난한 가족과 부유한 가족 사이 벌어지는 블랙코미디 풍자의 스릴러 영화."
    }
]
df = pd.DataFrame(data)
df.head(2)  # 데이터 확인 (실제로는 전체 데이터 사용)
```

```python
# 임베딩 할 문장 리스트 준비 (각 아이템의 synopsis 텍스트)
texts = df["synopsis"].tolist()
texts
```

```python
# OpenAI 임베딩 모델 설정 (text-embedding-3-small 사용)
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
texts=[]
metas=[]
ids=[]

for i,sample in enumerate(data):
  print(i,sample)

  text =sample['synopsis']
  metadata = {
    "title":sample['title'],
    "year":sample['year'],
    "genre":sample['genre'],
    "director":sample['director'],
    "actors":sample['actors'],
    "rating":sample['rating'],
    "text":sample['synopsis'],
  }
  texts.append(text)
  metas.append(metadata)
  ids.append(f"movie-{i}")

  print(ids)
  print(texts)
  print(metas)
  vectors = embeddings.embed_documents(texts)
  print(len(vectors[0]))
```

```python
movie_index.upsert(
    vectors=zip(ids,vectors,metas)
)
```

```python
query = "코미디 드라마 추천해줘"
results = vector_store.similarity_search(query,k=2)
for result in results:
  print(result)
```

```python
description = [movie['description'] for movie in movies]
movie_vectors = embeddings.embed_documents(description)
```

```python
pc.create_index(name="movie-index2",
                dimension=1536,
                metric="cosine",
                spec=ServerlessSpec(cloud="aws",
                                    region="us-east-1"))
```

```python
movie_index2 = pc.Index("movie-index2")
```

```python
vector_data=[]
for movie, vector in zip(movies,movie_vectors):
  metadata={
    "title":movie['title'],
    "year":movie['year'],
    "genre":movie['genre'],
    "description":movie['description']
  }
  vector_data.append((movie['id'],vector,metadata))
```

```python
movie_index2.upsert(vectors=vector_data)
```

```python
# query_text = '범죄와 관련된 영화 추천해줘.'
query_text = '2020년 이후에 개봉된 범죄와 관련된 영화 추천해줘.'
query_vector = embeddings.embed_query(query_text)
results = movie_index2.query(
  vector=query_vector,
  top_k=3,
  include_metadata=True
)

for result in results['matches']:
  print(f'Id : {result['id']}, Score: {result['score']}, Metadata: {result['metadata']} ')
```

```python
# query_text = '범죄와 관련된 영화 추천해줘.'
query_text = '2020년 이후에 개봉된 범죄와 관련된 영화 추천해줘.'
query_vector = embeddings.embed_query(query_text)
results = movie_index2.query(
  vector=query_vector,
  top_k=3,
  include_metadata=True,
  filter={
    "year":{"$gte":2020}
  }
)

for result in results['matches']:
  print(f'Id : {result['id']}, Score: {result['score']}, Metadata: {result['metadata']} ')
```
